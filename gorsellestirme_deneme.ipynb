
ax[0,0].bar_label(rects, padding=3) satırı, çubuk grafiğinde (bar chart) çubukların üzerine etiketler eklemek için kullanılır. Bu etiketler, genellikle çubuğun temsil ettiği sayısal değeri gösterir.

İşlem Aşamaları:

Grafiğin Çizilmesi:

    Çubuklar (bars) ax[0,0].bar() fonksiyonu ile çizilir. Bu işlem, veri değerlerini temsil eden dikdörtgen çubuklar oluşturur.

    ax[0,0].bar() fonksiyonunun çıktısı, her bir çubuğu temsil eden rects adlı Rectangle objelerinin listesidir.

Etiketlerin Eklenmesi:

    ax[0,0].bar_label(rects, padding=3) fonksiyonu, her çubuğun üzerine bir etiket ekler. 
    Bu etiketler, çubukların temsil ettiği sayısal değerleri gösterir.

    padding=3, etiketlerin çubuklardan yukarıda ne kadar uzaklıkta yerleştirileceğini belirler. Bu, etiketlerin çubukların üzerine yerleştirildiği aşamadır.

    Bu yüzden, ax[0,0].bar_label(rects, padding=3) satırı grafiğin çizilmesi değil, zaten çizilmiş olan çubuklara etiket ekleme aşamasıdır. Çubuklar çizildikten sonra, bu etiketler eklenir.x
