```python
import pandas as pd 
from IPython.display import HTML
from IPython.display import display
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_colwidth', None) 
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
    
def corner(home,away,hafta= 5,period= None): 
    df_all = pd.read_csv('/Users/icy/sofascore/sofa_calisma/csvler/last_df.csv', parse_dates=['date'])
    df = df_all[['date','org_detay','home_team','away_team','corner_kick_total','corner_kicks_home','corner_kicks_away','corner_kick_total_firsthalf','corner_kicks_home_firsthalf',
    'corner_kicks_away_firsthalf','corner_kick_total_secondhalf','corner_kicks_home_secondhalf','corner_kicks_away_secondhalf','home_goal','away_goal','result','home_iy_goal','away_iy_goal','iy_result','referee','season','ulke','season_round','season_period']]
    season = 7
    df  = df[~(df.ulke == 'Europe')]
    df_1 = df[(df.home_team.str.contains(home)) & (df.away_team.str.contains(away))].sort_values(by= 'date', ascending= False)
    df_2 = df[(df.home_team.str.contains(away)) & (df.away_team.str.contains(home))].sort_values(by= 'date', ascending= False)
    df_iki_takim_corner = pd.concat([df_1,df_2])
    toplam_corner = list(df_iki_takim_corner['corner_kicks_home'] + df_iki_takim_corner['corner_kicks_away'])
    toplam_iy_corner = list(df_iki_takim_corner['corner_kicks_home_firsthalf'] + df_iki_takim_corner['corner_kicks_away_firsthalf'])
    toplam_ikinci_corner = list((df_iki_takim_corner['corner_kicks_home'] + df_iki_takim_corner['corner_kicks_away']) - (df_iki_takim_corner['corner_kicks_home_firsthalf'] + df_iki_takim_corner['corner_kicks_away_firsthalf']))
    df_home_team_corner = df[(df.home_team.str.contains(home))].sort_values(by= 'date', ascending=  False).head(hafta)
    home_toplam_corner = list(df_home_team_corner['corner_kicks_home'] + df_home_team_corner['corner_kicks_away'])
    home_toplam_iy_corner = list(df_home_team_corner['corner_kicks_home_firsthalf'] + df_home_team_corner['corner_kicks_away_firsthalf'])
    home_toplam_ikinci_corner = list((df_home_team_corner['corner_kicks_home'] + df_home_team_corner['corner_kicks_away']) - (df_home_team_corner['corner_kicks_home_firsthalf'] + df_home_team_corner['corner_kicks_away_firsthalf']))
    df_away_team_corner = df[(df.away_team.str.contains(away))].sort_values(by= 'date', ascending=  False).head(hafta)
    away_toplam_corner = list(df_away_team_corner['corner_kicks_home'] + df_away_team_corner['corner_kicks_away'])
    away_toplam_iy_corner = list(df_away_team_corner['corner_kicks_home_firsthalf'] + df_away_team_corner['corner_kicks_away_firsthalf'])
    away_toplam_ikinci_corner = list((df_away_team_corner['corner_kicks_home'] + df_away_team_corner['corner_kicks_away']) - (df_away_team_corner['corner_kicks_home_firsthalf'] + df_away_team_corner['corner_kicks_away_firsthalf']))
    # if not df_1.empty:
    #     ulke = df_1.ulke.iloc[0]
    #     df_hakem_corner_ort = df[(df.ulke == ulke) & (df.season > season - 1)].groupby('referee')[['corner_kick_total','corner_kicks_home','corner_kicks_away','corner_kick_total_firsthalf','corner_kicks_home_firsthalf','corner_kicks_away_firsthalf']].mean().to_html()        
    #     df_hakem_corner_ort = f'<div style="max-height: 300px; overflow-y: auto;">{df_hakem_corner_ort}</div>'
    # else:
    #     df_hakem_corner_ort = pd.DataFrame()
    

    display(f'{home} - {away} Arasindaki Maclar')
    display(f'Toplam Korner = {toplam_corner} - Ilk Yari Toplam Korner = {toplam_iy_corner} - Ikinci Yari Toplam Korner = {toplam_ikinci_corner}')
    display(df_iki_takim_corner)
    display(f'{home} Son Ev Maclari Istatistik')
    display(f'Toplam Korner = {home_toplam_corner} - Ilk Yari Toplam Korner = {home_toplam_iy_corner} - Ikinci Yari Toplam Korner = {home_toplam_ikinci_corner}')
    display(df_home_team_corner)
    display(f'{away} Son Dep Maclari Istatistik')
    display(f'Toplam Korner = {away_toplam_corner} - Ilk Yari Toplam Korner = {away_toplam_iy_corner} - Ikinci Yari Toplam Korner = {away_toplam_ikinci_corner}')
    display(df_away_team_corner)
    if period is not None :
        df_home_period_corner =  df[(df.home_team.str.contains(home)) & (df.season_period == period)].sort_values(by= 'date', ascending=  False)
        df_away_period_corner =  df[(df.away_team.str.contains(away)) & (df.season_period == period)].sort_values(by= 'date', ascending=  False)
        df_home_period_corner_html =  df[(df.home_team.str.contains(home)) & (df.season_period == period)].sort_values(by= 'date', ascending=  False).to_html()
        df_away_period_corner_html =  df[(df.away_team.str.contains(away)) & (df.season_period == period)].sort_values(by= 'date', ascending=  False).to_html()
        df_home_period_corner_html = f'<div style="max-height: 300px; overflow-y: auto;">{df_home_period_corner_html}</div>'
        df_away_period_corner_html = f'<div style="max-height: 300px; overflow-y: auto;">{df_away_period_corner_html}</div>'
        home_toplam_corner_period = list(df_home_period_corner['corner_kicks_home'] + df_home_period_corner['corner_kicks_away'])
        home_toplam_iy_corner_period = list(df_home_period_corner['corner_kicks_home_firsthalf'] + df_home_period_corner['corner_kicks_away_firsthalf'])
        home_toplam_ikinci_corner_period = list((df_home_period_corner['corner_kicks_home'] + df_home_period_corner['corner_kicks_away']) - (df_home_period_corner['corner_kicks_home_firsthalf'] + df_home_period_corner['corner_kicks_away_firsthalf']))
        away_toplam_corner_period = list(df_away_period_corner['corner_kicks_home'] + df_away_period_corner['corner_kicks_away'])
        away_toplam_iy_corner_period = list(df_away_period_corner['corner_kicks_home_firsthalf'] + df_away_period_corner['corner_kicks_away_firsthalf'])
        away_toplam_ikinci_corner_period = list((df_away_period_corner['corner_kicks_home'] + df_away_period_corner['corner_kicks_away']) - (df_away_period_corner['corner_kicks_home_firsthalf'] + df_away_period_corner['corner_kicks_away_firsthalf']))
        display(f'{home} {period}. Periyot Korner Istatistik (Ev Sahibi Oldugu Maclar)')
        display(f'Toplam Korner =             {home_toplam_corner_period}')
        display(f'Ilk Yari Toplam Korner =    {home_toplam_iy_corner_period}')
        display(f'Ikinci Yari Toplam Korner = {home_toplam_ikinci_corner_period}')
        display(HTML(df_home_period_corner_html))
        display(f'{away} {period}. Periyot Korner Istatistik (Konuk Takim Oldugu Maclar)')
        display(f'Toplam Korner =             {away_toplam_corner_period}')
        display(f'Ilk Yari Toplam Korner =    {away_toplam_iy_corner_period}')
        display(f'Ikinci Yari Toplam Korner = {away_toplam_ikinci_corner_period}')
        display(HTML(df_away_period_corner_html))

    
```
